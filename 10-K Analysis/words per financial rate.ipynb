```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Dataframe data: contient toutes les infos/métriques + mda
# Dataframe counts: contient chaque mot et sa fréquence dans chaque document
data = pd.read_csv('final_labeled.csv')
```

```python
data_ind = pd.read_csv('data_industry.csv')
data['industry']=data_ind.gind.values
```

```python
data.dropna(inplace=True)
data.head()
```

```python
# Create counts from data 
vectorizer = CountVectorizer(stop_words="english")
counts = vectorizer.fit_transform(data.MDA.values)
```

```python
allWords = pd.DataFrame(sorted(vectorizer.vocabulary_.items(), key=lambda x: (x[1], x[0])),columns= ['word', 'wordID']).word.values
```

```python
good_count = pd.DataFrame(counts.toarray(),columns = allWords)
good_count.head()
```

```python
perfect_count = good_count.ix[:,50:]
perfect_count.head()
```

## Words vs RoE score (0 or 1)

```python
perfect_count['roe']=data.d_RoE_pos.values
table = pd.DataFrame(pd.pivot_table(perfect_count, index='roe', aggfunc=np.sum))
table.head()
```

```python
negative_roe = table.loc[0,:].sort_values(ascending=False)
print negative_roe
negative_roe.to_csv('negative_roe.csv')
print '###########################################'
positive_roe = table.loc[1,:].sort_values(ascending=False)
print positive_roe
positive_roe.to_csv('positive_roe.csv')
```

```python
k = 500
neg_words = pd.DataFrame(negative_roe).index.tolist()
#print neg_words[0:k]
pos_words = pd.DataFrame(positive_roe).index.tolist()
#print ''
#print pos_words[0:k]

print 'Words for negative RoE changes'
print [word for word in neg_words[0:k] if word not in pos_words[0:k]]
print ''
print 'Words for positive RoE changes'
print([word for word in pos_words[0:k] if word not in neg_words[0:k]])
```
